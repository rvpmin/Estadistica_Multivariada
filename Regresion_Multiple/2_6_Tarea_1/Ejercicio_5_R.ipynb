## a)
usar el algoritmo de selecicón hacia adelante para seleccionar un modelo de regresión.

```r
empty_model <- lm(y ~ 1, data = df)
```

```r
full_model <- lm(y ~ ., data = df)
```

```r
forward_model <- step(empty_model,
                      scope = formula(full_model),
                      direction = "forward")
```

1. y ~ x8
2. y ~ x8 + x2
3. y ~ x8 + x2 + x7
4. y ~ x8 + x2 + x7 + x9

## b)
Usar el algoritmo de selección hacia atrás para seleccionar un modelo de regresión.

```r
backward_model <- step(full_model, direction = "backward")
```

1. y ~ x1 + x2 + x3 + x4 + x5 + x6 + x7 + x8 + x9
2. y ~ x1 + x2 + x3 + x4 + x6 + x7 + x8 + x9
3. y ~ x2 + x3 + x4 + x6 + x7 + x8 + x9
4. y ~ x2 + x3 + x4 + x7 + x8 + x9
5. y ~ x2 + x4 + x7 + x8 + x9
6. y ~ x2 + x7 + x8 + x9 

## c)
Usar el algoritmo de regresión por pasos para seleccionar un modelo de regresión.

```r
stepwise_model <- step(empty_model, 
                        scope = formula(full_model), 
                        direction = "both")
```

1. y ~ x8
2. y ~ x8 + x2
3. y ~ x8 + x2 + x7
4. y ~ x8 + x2 + x7 + x9

## d)
Comenta los modelos finales en cada uno de los casos anteriores. ¿Cuál tiene más sentido? ¿Cuál modelo usarían?

```r
summary(forward_model)
```

```r
summary(backward_model)
```

```r
summary(stepwise_model)
```

Los tres métodos de selección de variables (*forward*, *backward* y *stepwise*) produjeron el mismo modelo final. Sin embargo, el método *stepwise* es el más robusto y con otros datos, el que podría dar mejores resultados.Las variables seleccionadas son consistentemente las más relevantes para explicar la variable dependiente y que el modelo es estable, por lo tanto este modelo es el más adecuado.
